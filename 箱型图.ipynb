```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']#解决中文乱码问题
plt.rcParams["font.size"]=10#设置字体大小
```

```python
data = pd.read_excel('./data/data_gt_60_new0901.xlsx')
# data = data[~((data['TG/HDL'] > 3.5) & ( data['mace'] ==0))]
```

```python
plt.boxplot([data.TyG指数[data.mace==1],data.TyG指数[data.mace==0]],labels=['mace','非mace'])
plt.title('TyG指数分布箱线图')
plt.show()
# plt.savefig('./pic/TyG箱线图.png',dpi=600)
```

```python
plt.boxplot([data[data.mace==1]['TG/HDL'],data[data.mace==0]['TG/HDL']],labels=['mace','非mace'])
plt.title('TG/HDL-C分布箱线图')
# plt.show()
plt.savefig('./pic/TGHDL-C箱线图.png',dpi=600)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt_data = data
plt_data['mace'] = data['mace'].apply(lambda x: "mace" if x else "非mace")
sns.set_style('darkgrid', {'font.sans-serif':['SimHei', 'Arial']})#设置图表背景颜色字体
sns.boxplot(y='TyG指数',data=plt_data,x='mace')
plt.title('TyG指数分布箱线图')
# plt.show()
plt.savefig('./pic/TyG指数箱线图-sb.png',dpi=600)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style('darkgrid', {'font.sans-serif':['SimHei', 'Arial']})#设置图表背景颜色字体
sns.boxplot(y='TG/HDL',data=plt_data,x='mace')
plt.title('TG/HDL-C箱线图')
# plt.show()
plt.savefig('./pic/TGHDL-C箱线图-sb.png',dpi=600)
```
