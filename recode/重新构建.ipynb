```python
stat_list = []
for col in ['合并心源性休克','合并心衰','合并心律失常', 
                          '合并恶性心律失常','室性心动过速','心室颤动', '房室传导阻滞', 
                          '严重房室传导阻滞','急性再发心肌梗死','冠状动脉支架内血栓形成', '急性亚急性支架血栓形成',
                         '合并死亡']:
    stat = data[(data['年龄']>60) & (data['mace']==True)][col].value_counts()
    print(stat)
    stat_list.append([col, stat.to_dict()])
```

```python
pd.DataFrame(stat_list,columns=["名称","统计值"]).to_excel('../data/0903/mace成因.xlsx',index=None)
```
